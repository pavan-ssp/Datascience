### Hypothesis testing : [ To find Interpretation and conclusion]

It is used in classification problem.

We try to evaluate two mutually exclusive statements on a population using sample data.

##Steps:

1) Make initial assumption as (H0).

2) Collect data {Evidences}.

3)Gather evidences to reject or accept null hypothesis.

4) If we want to reject null hypothesis we can accept alternative hypothesis (H1).



####Type-1 Error :     

Due to the lack of evidence we need to reject the null hypothesis and accept the alternative hypothesis.

###Type-2 Error :

H0 market is going to crash Dont have enough evidence H1 is ignored and H0 is taken.





###Chisquare Test :     

If we are using 2 categorical features and we judge H1 or H0 Hypothesis then we use ""Chisquare-Test"

### One Sample Proportion :

If we have 1 categorical feature and then we use "One Sample Proportion" Test.


### T-Test :      

If we have one numerical variable feature then we use "T-Test".


### Co-relation :     

If we have 2 Numerical variable features then we use "Co-relation".

### Annova Test :     

If we have 1 numerical and more than 1 categorical variable then we use "Annova-Test".

##Recall :      

Recall = Tp / (TP+FN)

Out of all total positive actual values how many positive values we predicted correctly.

Eg :    

If a person having cancer but the model is not  predicted as cancer.
