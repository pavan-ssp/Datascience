###MultiCollinearity in Regression

--> Here we find the collinearity between the features and their collinearity percentages.

--> Collinearity is the percentage of features which will depend on other which we are removed so that it will not impact the model.

--> We add constant to the model and find the summary of the model such that the standarderror is seen and P value is observed such that the standarderror should be low and corresponding P value should less than the 0.5 so that the conclusion that the dependent on the other feature is low.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('/content/drive/MyDrive/Classroom/Salary_Data.csv')
df
```

```python
import statsmodels.api as sm
```

```python
x=df[['YearsExperience','Age']]
y=df['Salary']
```

```python
x
```

```python
x=sm.add_constant(x)  # Here we added one column constant such that all  values are 1.0
x
```

```python
model=sm.OLS(y,x).fit() # Here first value we need to pass y and then x and fit()
```

```python
model.summary()
```

Here first we see r2 that is fine

Now we go to standard error is too high so that we may assume that it has some collinearity between the features.

Now we go to p value for age the p values is high greater than of .5.

So if we want to remove one feature we  can remove the high p value such that we can overcome  collinearity between them.

Further we want to check collinearity we use collinearity matrix.

```python
x.iloc[:,1:].corr()
```

###conclusion

Here we can see that 98% corelation between the features we can drop the age column such that it contains high p value to optimize the model.
